6. Hãy thực hiện xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại.

```python
def fill(columns):
  return dlxtdh[columns].fillna(dlxtdh[columns].mean(), inplace = True)
is_float64 = dlxtdh.dtypes == float
dlxtdh_column_list = dlxtdh.columns[is_float64].tolist()
for col in dlxtdh_column_list:
    dlxtdh[col].fillna(dlxtdh[col].mean(), inplace = True)
dlxtdh.head()
```

7. Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và 12. • Công thức tính: TBM = (T2 + L + H + S + V2 + X + D + N) / 10

```python
for i in ["1","2","6"]:
    dlxtdh["TBM"+i] = (dlxtdh["T"+ i]*2 + dlxtdh["L" + i] + dlxtdh["H" + i] + dlxtdh["S" + i] + dlxtdh["V" + i]*2 + dlxtdh["X" + i] + dlxtdh["D" + i] + dlxtdh["N" + i]) / 10
dlxtdh.rename(columns = {"TBM6": "TBM3"}, inplace = True)
dlxtdh
```

8. Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1, TBM2 và TBM3 cho từng năm lớp 10, 11, 12 như sau: • Nhỏ hơn 5.0 xếp loại: yếu (kí hiệu là Y) • Từ 5.0 đến dưới 6.5: trung bình (kí hiệu là TB) • Từ 6.5 đến dưới 8.0: khá (kí hiệu là K) • Từ 8.0 đến dưới 9.0: giỏi (kí hiệu là G) • Từ 9.0 trở lên: xuất sắc (kí hiệu là XS)

```python
for i in ["1","2","3"]:
    dlxtdh.loc[dlxtdh["TBM" + i] < 5, "XL" + i] = "Y"
    dlxtdh.loc[(dlxtdh["TBM" + i] >= 5) & (dlxtdh["TBM" + i] < 6.5), "XL" + i] = "TB" 
    dlxtdh.loc[(dlxtdh["TBM" + i] >= 6.5) & (dlxtdh["TBM" + i] < 8), "XL" + i] = "K" 
    dlxtdh.loc[(dlxtdh["TBM" + i] >= 8) & (dlxtdh["TBM" + i] < 9), "XL" + i] = "G"
    dlxtdh.loc[(dlxtdh["TBM" + i] >= 9), "XL" + i] = "XS"
dlxtdh
```

9. Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp 10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương pháp Min-Max Normalization

```python
for i in ["1", "2", "3"]:
  dlxtdh["US_TBM" + i] = dlxtdh["TBM" + i] / 10 * 4
dlxtdh
```

10. Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định sinh viên đậu (giá trị “1”) và rớt (giá trị “0”) vào các khối dựa trên điểm DH1, DH2 và DH3 như sau • Với khối A, A1 nếu [(DH12 + DH2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt • Với khối B nếu [(DH1 + DH22 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt • Với khối khác nếu [(DH1+ DH2 + DH3)/3] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt

```python
for i in  dlxtdh["KT"].unique():
  dlxtdh.loc[(i == "A1" or i == "A") & (((dlxtdh["DH1"]*2 + dlxtdh["DH2"] + dlxtdh["DH3"])/4) >= 5), "KQXT"] = 'Đậu'
  dlxtdh.loc[(i == "A1" or i == "A") & (((dlxtdh["DH1"]*2 + dlxtdh["DH2"] + dlxtdh["DH3"])/4) < 5), "KQXT"] = 'Rớt'
    
  dlxtdh.loc[(i == "B") & (((dlxtdh["DH1"] + dlxtdh["DH2"]*2 + dlxtdh["DH3"])/4) >= 5), "KQXT"] = 'Đậu'
  dlxtdh.loc[(i == "B") & (((dlxtdh["DH1"] + dlxtdh["DH2"]*2 + dlxtdh["DH3"])/4) < 5), "KQXT"] = 'Rớt'
    
  dlxtdh.loc[(i != "A1" and i != "A" and i != "B") & (((dlxtdh["DH1"] + dlxtdh["DH2"] + dlxtdh["DH3"])/3) >= 5), "KQXT"] = 'Đậu'
  dlxtdh.loc[(i != "A1" and i != "A" and i != "B") & (((dlxtdh["DH1"] + dlxtdh["DH2"] + dlxtdh["DH3"])/3) < 5), "KQXT"] = 'Rớt'
dlxtdh
```
